## Turn Each Column to Clustered_Data
with Mean Shift

Output Shows Range of each Cluster as well

```python

clustered_Data =[] 
labels=[]
cluster_centers=[]
labels_unique=[]
n_clusters_=[]
ranges_column=[]

for i in range(2,np_record.shape[1]):
    X=np_record[:,i].reshape(np_record[:,i].shape[0],1)

    X=X.astype('float64')
    if np.isnan(X).any() == True:
        print("There is NAN")
        X = np.nan_to_num(X)

    ms = MeanShift(bandwidth=None, bin_seeding=True)
    ms.fit(X)

    labels.append(ms.labels_)

    cluster_centers.append(ms.cluster_centers_)

    

    labels_unique.append(np.unique(ms.labels_))

    n_clusters_.append(len(np.unique(ms.labels_)))

 

    print("number of estimated clusters : %d" % len(np.unique(ms.labels_)))

    print(ms.labels_)
    clustered_Data.append(ms.labels_)

    arr=[]
    for k in np.unique(ms.labels_):
        print([min(X[ms.labels_==k]),max(X[ms.labels_==k])])
        arr.append([min(X[ms.labels_==k]),max(X[ms.labels_==k])])
    ranges_column.append(arr)
```

```python

association_rules = apriori(cls_data, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=1)
association_results = list(association_rules)
```
